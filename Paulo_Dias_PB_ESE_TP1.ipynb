# Ponto de partida

### Primeira Entrega: Construção de um Monólito Simples com Spring Boot

**Objetivo da Etapa:**

* Nesta primeira etapa, os alunos desenvolverão uma aplicação monolítica simples utilizando Spring Boot, preparando a base para futuras evoluções em direção a uma arquitetura de microsserviços.


**Subcompetências a serem Desenvolvidas:**

* **Implementação de uma Aplicação com Spring Boot:**
    * Desenvolver uma aplicação seguindo um design em camadas (controle, serviço e repositório).
    * Utilizar a Autoconfiguração do Spring Boot para simplificar o setup inicial do projeto.


* **Gerenciamento de Dependências:**
    * Aprender a gerenciar dependências de forma eficaz, preparando o projeto para escalabilidade e manutenção.

* **Desenvolvimento de APIs REST:**
    * Implementar APIs REST utilizando as anotações do Spring MVC para facilitar a comunicação entre o front-end e o back-end.

* **Práticas de Código Limpo e Manutenção:**
    * Aplicar princípios SOLID e padrões de design para garantir que o código seja limpo e fácil de manter.

* **Modelagem de Domínios:**
    * Modelar casos de negócio usando conceitos de domínios, subdomínios e bounded contexts do Domain-Driven Design (DDD).

* **Desenvolvimento Front-End:**
    * Criar uma interface de usuário básica usando React para consumir informações do back-end, facilitando a interação do usuário com a aplicação.

**Entregas Esperadas:**
* Código Fonte: Repositório Git com o código fonte da aplicação, incluindo front-end e back-end.
* Documentação: Documento explicando a arquitetura da solução, incluindo diagramas de componentes e sequência que ilustrem o design da aplicação.
* Demonstração: Uma apresentação prática da aplicação em funcionamento, demonstrando as funcionalidades implementadas e a interação entre front-end e back-end.

**Avaliação:**
* A avaliação será baseada na completude da implementação, qualidade do código, aderência aos requisitos de design e arquitetura, e na eficácia da apresentação e documentação entregue.

# Resolução

Primeiramente foi criado uma aplicação Spring Boot utilizando da plataforma Spring Initializr conforme a imagem abaixo.



Para este projeto, foi idealizado uma estrutura de "Fale comigo", onde o usuário adiciona nome, e-mail e uma mensagem ao proprietário do site, uma forma de captar revies, solicitações, etc. Foram criados a estrutura conforme o print abaixo:



Repository > ContactRepository: Responsável por extender o repositório JPA JpaRepository, atuando como um repositório para uso da aplicação;

Service > ContactServise: Responsável pela implementação de respositório, criando dois métodos, save() e listAll();

Facade > ContactFacade: Camada intermediária entre o ContactServise e o ContactController. Atua com a criação de dois métodos createContact que chama save, e listContact que chama listAll();

Controller > ContactController: Responsável pela implementação dos END POINTS. Implementado pensando na implementação do Front-End futuramente, possui anotações @RestController, @RequestMapping("/contact") e @CrossOriging(origing = "*"). Possui implementação de create() - Post - e listAll() - Get.

Models > ContactModel: Classe responsável pela modelagem da entidade. PS: Não consegui configurar corretamente o Lombok, então para Getters e Setters realizei da forma regular, irei corrigir isso na evolução da entrega.

Melhorias serão implementadas ao decorrer da evolução do projeto.
